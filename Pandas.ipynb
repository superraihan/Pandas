### Sorting

```python
import pandas as pd

df = pd.read_csv('student_data.csv')
print(df)
```

```python
df
```

```python
# Low To High Marks

copy = df.sort_values('Data Structure Marks')
copy
```

```python
# High To Low Marks

copy = df.sort_values('Data Structure Marks',ascending=False)
copy
```

```python
# Mixed Marks High To Low 

copy = df.sort_values(['Data Structure Marks', 'Python Marks'],ascending=False)
copy
```

```python
# Mixed Marks Low To High

copy = df.sort_values(['Data Structure Marks', 'Python Marks'],ascending=True) #[0,1] Use Kora Jete Pare True False Er Bolode
copy
```

### Filtering Data based on condition

```python
df
```

```python
# Find Value Who Not Started Course

not_started = df.loc[df['CompletionStatus'] =='Not Started']
not_started
```

```python
# Find Value Who Completed Course

Completed_Courses = df.loc[df['CompletionStatus'] =='Completed']
Completed_Courses
```

```python
# Completed And DS Marks Is Getter Or Equal 90

Completed_ds90 = df.loc[(df['CompletionStatus'] =='Completed') & (df['Data Structure Marks']>=90)]

Completed_ds90
```
